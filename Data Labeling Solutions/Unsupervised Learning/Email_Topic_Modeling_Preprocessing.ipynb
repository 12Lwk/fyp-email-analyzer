```python
from transformers import pipeline

summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_text(text):
    if len(text.split()) > 512:
        return summarizer(text, max_length=150, min_length=50, do_sample=False)[0]['summary_text']
    return text

email_data["Summarized_Text"] = email_data["Combined_Text"].apply(summarize_text)
```

```python
from bertopic import BERTopic

topic_model = BERTopic()
topics, probs = topic_model.fit_transform(email_data["Summarized_Text"])

email_data["Topic"] = topics
topic_model.get_topic_info()
```

```python
topic_model.get_topic_info()
```

```python
topic_to_category = {
    0: "Promotional",
    1: "Important",
    2: "Spam",
    3: "Finance",
    4: "General",
}

email_data["Category"] = email_data["Topic"].map(topic_to_category).fillna("General")
```

```python
def assign_priority(text):
    text = text.lower()
    if any(word in text for word in ["urgent", "asap", "important", "deadline"]):
        return "High"
    elif any(word in text for word in ["reminder", "follow-up", "schedule"]):
        return "Medium"
    else:
        return "Low"

email_data["Priority"] = email_data["Summarized_Text"].apply(assign_priority)
```
